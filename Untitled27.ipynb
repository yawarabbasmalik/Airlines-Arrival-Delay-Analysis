```python
import pandas as pd

# Load the dataset
file_path = '/content/Invistico_Airline.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
# Checking for null values
null_values = df.isnull().sum()
null_values
```

```python
# Checking for duplicate values
duplicate_values = df.duplicated().sum()
duplicate_values
```

```python
# Removing the missing values and duplicates from the dataframe
df_cleaned = df.dropna().drop_duplicates()

# Checking the shape of the cleaned data to confirm the changes
cleaned_shape_of_data = df_cleaned.shape

cleaned_shape_of_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Preparing for EDA focusing on 'satisfaction'
# Selecting a few relevant columns for analysis
columns_for_eda = ['Age', 'Flight Distance', 'Inflight wifi service',
                   'Ease of Online booking', 'On-board service', 'Seat comfort',
                   'Inflight entertainment', 'Cleanliness', 'satisfaction']

# Creating a subset of the dataframe for EDA
df_eda = df_cleaned[columns_for_eda]

# Creating visuals

# Visual 1: Satisfaction Count
plt.figure(figsize=(8, 5))
sns.countplot(x='satisfaction', data=df_eda)
plt.title('Count of Customer Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Count')
plt.show()
```

```python
# Visual 2: Satisfaction by Age
plt.figure(figsize=(12, 6))
sns.boxplot(x='satisfaction', y='Age', data=df_eda)
plt.title('Customer Satisfaction by Age')
plt.xlabel('Satisfaction')
plt.ylabel('Age')
plt.show()
```

```python
# Visual 3: Satisfaction by Flight Distance
plt.figure(figsize=(12, 6))
sns.boxplot(x='satisfaction', y='Flight Distance', data=df_eda)
plt.title('Customer Satisfaction by Flight Distance')
plt.xlabel('Satisfaction')
plt.ylabel('Flight Distance')
plt.show()
```

```python
# Visual 4: Satisfaction by Inflight Entertainment
plt.figure(figsize=(12, 6))
sns.boxplot(x='satisfaction', y='Inflight entertainment', data=df_eda)
plt.title('Customer Satisfaction by Inflight Entertainment')
plt.xlabel('Satisfaction')
plt.ylabel('Inflight Entertainment Rating')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Encoding categorical columns using Label Encoding
label_encoder = LabelEncoder()
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
df_encoded = df_cleaned.copy()

for column in categorical_columns:
    df_encoded[column] = label_encoder.fit_transform(df_encoded[column])

# Performing data normalization using Min-Max Scaling
min_max_scaler = MinMaxScaler()
df_normalized = pd.DataFrame(min_max_scaler.fit_transform(df_encoded), columns=df_encoded.columns)

df_normalized.head()
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Defining the models
models = {
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Neural Networks": MLPClassifier(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

# Target variable and features
X = df_normalized.drop('satisfaction', axis=1)
y = df_normalized['satisfaction']

# Stratified K-Fold for maintaining the proportion of target class in each fold
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Evaluation metrics
scoring_metrics = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score)
}

# Performing 10-fold cross-validation for each model and each metric
results = {}
for model_name, model in models.items():
    results[model_name] = {}
    for metric_name, metric in scoring_metrics.items():
        scores = cross_val_score(model, X, y, cv=cv, scoring=metric)
        results[model_name][metric_name] = scores.mean()

results
```
